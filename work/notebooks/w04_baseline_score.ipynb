## 1. My rule and its reason codes

Rule (plain words):
Rank keywords higher if they have high search volume and low competition.
Lower rank if competition is high or search volume is low.

Scoring idea:
Action Score = (search_volume normalized) − (competition)

Reason Codes:

R1: High Volume, Low Competition → strong opportunity
R2: High Volume, High Competition → risky but valuable
R3: Low Volume, Low Competition → niche opportunity
R4: Low Volume, High Competition → weak candidate

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import pandas as pd
import numpy as np
import os

# Sample dataset (replace with your real dataset)
data = {
    'query': ['shoes', 'running shoes', 'phone', 'best phone', 'laptop',
              'gaming laptop', 'cheap shoes', 'iphone', 'android phone', 'headphones',
              'wireless headphones', 'earbuds', 'smartwatch', 'tablet', 'camera',
              'dslr camera', 'tripod', 'monitor', 'keyboard', 'mouse'],
    'search_volume': [1000, 800, 1500, 1200, 900, 700, 600, 1400, 1100, 500,
                      650, 400, 300, 350, 450, 550, 200, 750, 650, 600],
    'competition': [0.8, 0.6, 0.9, 0.7, 0.5, 0.6, 0.4, 0.85, 0.75, 0.3,
                    0.4, 0.2, 0.3, 0.35, 0.5, 0.55, 0.2, 0.6, 0.5, 0.45]
}

df = pd.DataFrame(data)

# Normalize search volume
df['norm_volume'] = df['search_volume'] / df['search_volume'].max()

# Baseline action score
df['action_score'] = df['norm_volume'] - df['competition']

# Ranking
df = df.sort_values(by='action_score', ascending=False)

# Save CSV
os.makedirs("work/outputs", exist_ok=True)
df.to_csv("work/outputs/baseline_action_score.csv", index=False)

print("Saved to work/outputs/baseline_action_score.csv")
df.head(10)
```
