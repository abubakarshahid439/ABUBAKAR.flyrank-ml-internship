## 1. Ranked actions + reason codes

Ranked Content Actions (Observed & Directional)

Update high-impression, low-CTR content
Reason: Observed that pages with high impressions but low CTR may have weak titles or meta descriptions.
Action: Improve headlines and meta descriptions.
Refresh outdated content with declining traffic
Reason: Measured traffic drop suggests content may be outdated.
Action: Update statistics, keywords, and examples.
Expand high-performing content
Reason: Pages with high engagement show strong user interest.
Action: Add more depth, FAQs, and internal links.
Fix low-engagement content
Reason: Low scroll rate and engagement indicate weak content quality.
Action: Improve readability and structure.
Remove or merge underperforming pages
Reason: Very low impressions and engagement suggest low value.
Action: Merge with related content or delete

```python
import pandas as pd

# Example dataset (replace with your real data)
data = {
    "content_id": ["A", "B", "C", "D", "E"],
    "impressions": [10000, 5000, 8000, 2000, 500],
    "ctr": [0.01, 0.05, 0.02, 0.03, 0.005],
    "engagement": [0.2, 0.6, 0.5, 0.3, 0.1]
}

df = pd.DataFrame(data)

# Ranking logic
df["priority_score"] = df["impressions"] * (1 - df["ctr"])

df = df.sort_values(by="priority_score", ascending=False)

df.head()
```
