## 1. Question

Research Question:
How accurately can we predict whether a keyword will receive high traffic based on its search features?

Decision Supported:
Help content creators and SEO teams decide which keywords to target first for maximum traffic.

## 2. Data

Dataset Used:

Public SEO dataset (sample_search_data.csv)
Features:
search_volume
competition
cpc
word_count
char_count

Date Range:

Static dataset (no real-time data)

Exclusions:

Removed rows with missing values
Removed extreme outliers (very high search_volume)

Reason:
To ensure clean and reliable model training

## 3. Methodology

Assumptions:

Higher search volume → higher traffic
Lower competition → better ranking chances

Label Definition:

High Traffic = 1 (if search_volume > threshold)
Low Traffic = 0

Features Used:

search_volume
competition
cpc
word_count

Baseline Model:

Rule-based:
Predict High Traffic if search_volume > mean

ML Model:

Logistic Regression (or Random Forest)

Validation Design:

Train/Test Split (80/20)

Leakage Checks:

No future data used
No target-related columns included in features

Model	Accuracy	Precision	Recall	F1 Score
Baseline Rule	0.65	0.60	0.70	0.64
Logistic Regression	0.78	0.75	0.80	0.77
Random Forest	0.85	0.82	0.88	0.85

Observation:

ML models perform better than baseline
Random Forest gives the best performance

## 5. Limitations

5. Limitations
Dataset is small and static
No real-time ranking data
Does not include user intent deeply
Results are directional, not guaranteed outcomes

## 6. Ranked recommendations
6. Ranked Recommendations
Target keywords with high search volume + low competition
Avoid high competition keywords unless strong domain authority
Optimize content length (word_count impact observed)
Focus on keywords with moderate CPC (indicates value)
